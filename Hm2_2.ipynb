```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("file.csv")

print(df.head())

print("Розмір таблиці:", df.shape)

print("Типи всіх стовпців:")
print(df.dtypes)

missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Частка пропусків міститься в кожній колонці:")
print(missing_percentage)

columns_to_drop = df.columns[df.isnull().any() & (df.columns != 'Мова програмування')]

df = df.drop(columns_to_drop, axis=1)

print("Частка пропусків міститься в кожній колонці після видалення:")
print((df.isnull().sum() / len(df)) * 100)

df = df.dropna()

print("Розмір нової таблиці:", df.shape)

python_data = df[df['Мова програмування'] == 'Python']

print("Розмір таблиці python_data:", python_data.shape)

grouped_data = df.groupby('Посада')

aggregated_data = grouped_data['salary'].agg(['min', 'max'])

def fill_avg_salary(salary):
    return salary

df['avg'] = df['Зарплата на місяць'].apply(fill_avg_salary)

print(df['avg'].describe())

df.to_csv('output.csv', index=False)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Зарплата на місяць'], kde=True, color='skyblue', bins=30)
plt.title('Розподіл зарплат за місяць')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Кількість спостережень')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Посада', data=df, palette='viridis')
plt.title('Кількість працівників за посадами')
plt.xlabel('Посада')
plt.ylabel('Кількість працівників')
plt.xticks(rotation=45)
plt.show()



plt.figure(figsize=(10, 6))
sns.boxplot(x='Мова програмування', y='Зарплата на місяць', data=df)
plt.title('Порівняння зарплат за мовами програмування')
plt.xlabel('Мова програмування')
plt.ylabel('Зарплата')
plt.xticks(rotation=45)
plt.show()

```
